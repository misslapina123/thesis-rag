```python
difference_values = []

# Iterate through each dataframe
for df in dataframes:
    # Find the maximum value in the "Average" column
    max_avg_value = df["Average"].max()
    # Find the average value corresponding to GPT-3.5
    avg_gpt35 = float(df[df["System"] == "GPT-3.5"]["Average"])
    # Calculate the difference and append it to the list
    difference_values.append(max_avg_value - avg_gpt35)

# # Calculate the average of all the difference values
average_difference = sum(difference_values) / len(difference_values)

print("Average difference between max value and average value of GPT-3.5 across all dataframes:", average_difference)
```
