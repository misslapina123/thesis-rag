```python
difference_values = []

# Iterate through each dataframe
for df in dataframes:
    # Find the maximum value in the "Average" column for systems containing "GPT-4"
    max_avg_value_gpt4 = df[df["System"].str.contains("GPT-4")]["Average"].max()
    
    # Find the maximum value in the "Average" column for the regular "GPT-4" system
    max_avg_value_regular_gpt4 = df[df["System"] == "GPT-4"]["Average"].max()

    # Compare and choose the greater value between the two
    max_avg_value = max(max_avg_value_gpt4, max_avg_value_regular_gpt4)

    # Find the average value corresponding to GPT-4
    avg_gpt4 = float(df[df["System"].str.contains("GPT-4")]["Average"].mean())

    # Calculate the difference and append it to the list
    difference_values.append(max_avg_value - avg_gpt4)

# Calculate the average of all the difference values
average_difference = sum(difference_values) / len(difference_values)

print("Average difference between max value and average value of GPT-4 across all dataframes:", average_difference)
```
