```python
import os
import pandas as pd

# Path to the folder containing the CSV files
input_folder = r"C:\Users\sigit\OneDrive\TSI\thesis\thesis-rag\ballad\results"

# Path to the folder where you want to save the results
output_folder = r"C:\Users\sigit\OneDrive\TSI\thesis\thesis-rag\results"

# Iterate over each file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".csv"):
        # Read the CSV file
        filepath = os.path.join(input_folder, filename)
        df = pd.read_csv(filepath)
        
        # Find the top 3 rows with the highest values in the "Average" column
        top_3_average = df.nlargest(3, 'Average')

        # Initialize an empty DataFrame to store the final result
        final_result = pd.DataFrame()

        # Iterate over the columns (excluding 'Average' and 'System')
        for column in df.columns.difference(['Average', 'System']):
            # Find the row with the highest value for the current column among the top 3 rows with highest average
            top_1_in_column = df.nlargest(1, column)
            # Check if the index of the top 1 in the column is not already in the final result
            if top_1_in_column.index[0] not in final_result.index:
                # Append the row to the final result DataFrame
                final_result = final_result.append(top_1_in_column)

        # Concatenate the final result with the top 3 rows with highest average
        final_result = pd.concat([final_result, top_3_average])
        
        # Filter rows where System is either 'GPT-3.5' or 'GPT-4'
        gpt_3_5_4_rows = df[df['System'].isin(['GPT-3.5', 'GPT-4'])]

        # Concatenate these rows with the final result
        final_result = pd.concat([gpt_3_5_4_rows, final_result])
        
        # Remove duplicates
        final_result = final_result.drop_duplicates()

        # Extract type from the filename
        file_type = filename.split('.')[0].split('_')[-1]
        
        # Construct the output filepath
        output_filepath = os.path.join(output_folder, f"results_ballad_{file_type}.csv")

        # Save the final result to CSV
        final_result.to_csv(output_filepath, index=False)
```
